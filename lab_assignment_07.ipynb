```python
#imported required libraries
import pandas as pd
#import portfolio data
pf_data = pd.read_csv("/content/gdrive/MyDrive/portfolio_data.csv")
pf_data
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
#Prediction using linear regression model
x = pf_data.drop(['Date', 'NFLX'], axis = 1)
y = pf_data['NFLX']

#splitting the data into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 40)

#predictions based on linear regression model
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
#predictions on the test set
y_predict_test = linear_reg.predict(X_test)
print("RMSE score in linear regression model on test data : ",np.sqrt(mean_squared_error(y_test, y_predict_test)))
print("R2 Score in linear regression model on test data : ", r2_score(y_test, y_predict_test))
```

```python
#Prediction using non-linear regression model
#using decision tree regressor as it is one of the non linear regression model
dec_tree = DecisionTreeRegressor(max_depth=8, min_samples_leaf=0.13, random_state=3)
#fitting the model on the training set 
dec_tree.fit(X_train, y_train)
#prediction based on test data 
predict_tree_test= dec_tree.predict(X_test)
#prints the evaluation metrics - RMSE and R-squared - on the test set.
print("RMSE score in decision tree regression model on test data", (np.sqrt(mean_squared_error(y_test,predict_tree_test))))
print("R2 score in decision tree regression model on test data",r2_score(y_test, predict_tree_test))
```

```python
#prediction based on training data 
predict_tree_train= dec_tree.predict(X_train)
#prints the evaluation metrics - RMSE and R-squared - on the training set.
print("RMSE score in decision tree regression model on train data", (np.sqrt(mean_squared_error(y_train,predict_tree_train))))
print("R2 score in decision tree regression model on train data",r2_score(y_train, predict_tree_train))
```

### Question 2 (25 points) The objective of this task is to detect hate speech in tweets. For the sake of simplicity, we say a tweet contains hate speech if it has a racist or sexist sentiment associated with it. So, the task is to classify racist or sexist tweets from other tweets. Formally, given a training sample of tweets and labels, where label '1' denotes the tweet is racist/sexist and label '0' denotes the tweet is not racist/sexist, your objective is to predict the labels on the test dataset. The dataset can be downloaded here: https://github.com/unt-iialab/info5502-spring2022/tree/main/datasets/Twitter%20Sentiment%20Analysis. Please build a logistic regression model for the classification, evaluation metrics include accuracy, precision, recall, and F-1 score need to be reported. 

```python
### You code here:

#importing necessary libraries
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import chi2
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
import pickle
import sqlite3
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('stopwords')
#stop words of englis language
stopwords = stopwords.words("english")
stopwords = [re.sub('[^a-zA-Z0-9]+', '', word) for word in stopwords]
```

```python
#imported the train dataset 
train_data = pd.read_csv("/content/gdrive/MyDrive/train.csv")
train_data
```

```python
#removing all the special characters from the tweets and cleansing the column tweets
train_data['tweet_cleansed'] = train_data['tweet'].str.replace('[^a-zA-Z0-9 ]', '')
train_data['tweet_cleansed'] = train_data['tweet_cleansed'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
#getting the data where label=1 which are racist/sexist tweets
racist_tdata = train_data[train_data.label == 1]
racist_tdata
```

```python
regular_tdata= train_data[train_data.label == 0]
regular_tdata
```

```python

nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()

# applying lemmatization on the tweets
train_data['tweet_cleansed'] = train_data['tweet_cleansed'].apply(lambda row: " ".join(wordnet_lemmatizer.lemmatize(word, pos="v") for word in row.split()))
train_data['tweet_cleansed'] = train_data['tweet_cleansed'].apply(lambda row: " ".join(wordnet_lemmatizer.lemmatize(word, pos="a") for word in row.split()))
train_data['tweet_cleansed'] = train_data['tweet_cleansed'].apply(lambda row: " ".join(wordnet_lemmatizer.lemmatize(word, pos="n") for word in row.split()))
train_data['tweet_cleansed'] = train_data['tweet_cleansed'].apply(lambda row: " ".join(wordnet_lemmatizer.lemmatize(word, pos="r") for word in row.split()))
```

```python
X = train_data['tweet_cleansed']
Y = train_data['label']
# Performs the TF-IDF transformation
vect = TfidfVectorizer(min_df = 3, ngram_range=(1, 2))
#fit the train data and generating the required features
vect = vect.fit(X)
X_char = vect.transform(X).toarray()
X_train, X_test, Y_train, Y_test = train_test_split(X_char, Y, test_size=0.4)

# building logistic regression model based on the given data 
logistic_reg = LogisticRegression()
pred = logistic_reg.fit(X_train, Y_train)
```

```python
# printing the evaluation metrics which includes precision,recall,f1 score and support
print("Confusion matrix obtained:")
print(confusion_matrix(Y_test, pred.predict(X_test)))
print("Classification report obtained:")
print(classification_report(Y_test, pred.predict(X_test)))
```
